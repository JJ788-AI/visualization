```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

np.random.seed(613)
x = np. arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
#Adding lable 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1,label = 'Person 1')
ax.plot(x,y2,label = 'Person 2')
ax.legend(loc= 'lower right',
          frameon = True,
          fontsize =12,
          ncol =2,        #label in 2 columns 
          shadow =True)

```

```python
#moving the legend outside 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y1,label = "Person 1")
ax.plot(x,y2,label = "Person 2")
ax.legend(loc= "upper left",
          bbox_to_anchor=(1,1))
```

```python
#text and annotations
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1,label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc= "lower right")
ax.text(10,95, "This value is important!",
        ha ="center",
        color ="red",
        size =20)
##modifying text annotations ##
```

```python
#position text annotations 
fig, ax =plt.subplots()
ax.axis([0,10,0,10])
#position annotation with reference to the data
ax.text(1,5, ". Data: (1,5)", transform =ax.transData)
ax.text(0.5,0.1 ,". Axes:(0.5, 0.1)", transform = ax.transAxes)
ax.text(0.2,0.2, ". Figure:(0.2,0.2)", transform=fig.transFigure)
```

```python
#annotation with arrows
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1,label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc= "lower right")
ax.annotate("This is important!", xy= (10,95),
            xytext=(20,94),
            arrowprops=dict(facecolor ='black', shrink=0.05))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1,label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc= "lower right")
ax.annotate("This is important!", xy= (10,95),
            xytext=(20,94),
            arrowprops=dict(facecolor ='hotpink', arrowstyle = "wedge"))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1,label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc= "lower right")

ax.yaxis.set_major_locator(plt.NullLocator())
#locator to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter())
#formatter to remove only labels
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO
```

```python
x=np.array(["Luffy", "Zoro", "Nami","Usopp","Sanji"])
y1=np.array([110,180,240,99,220])
y2=np.array([170,100,90,120,50])

#define our figure and axes
fig, ax =plt.subplots(figsize=(7,3))
ax.bar(x,y1,
       color="indigo")
ax.plot(x,y2,
        color="red")
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color ="red")
ax.errorbar(x,y2,
            yerr=y2_sd,
            fmt="none")
```

```python
response= requests.get ('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file=BytesIO(response.content)
image=Image.open(image_file)
```

```python
fig, ax =plt.subplots(figsize=(7,3))
ax.plot(x,y2,color="red")
ax_image=fig.add_axes([0.1,0.11,0.15,0.35])
ax_image.imshow(image)
ax_image.axis('off')
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('whitegrid')
sns.lineplot(data=tips,
x='total_bill',
y='tip')
```

```python
tipgraph = sns.lineplot(data=tips,
                       x ='total_bill',
                       y ='tip')
tipgraph.set(title='Tips vs Total Bill',
             xlabel ='Total Bill($)',
             ylabel ='Tip Amount($)')
```

```python
fig = plt.subplots(figsize=(10,3))

tipgraph = sns.lineplot(data=tips,
                        x= 'total_bill',
                        y='tip',
                        color='hotpink',
                        linestyle ='--',
                        linewidth=3,
                        marker ='o',
                        markerfacecolor='indigo')

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill($)',
             ylabel='Tip Amount($)')
```

```python
tipgraph = sns.scatterplot(data=tips,
                        x= 'total_bill',
                        y='tip',
                        style='time',
                        hue='day',
                        palette=['purple','hotpink','deepskyblue','yellowgreen'])

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill($)',
             ylabel='Tip Amount($)')
```

```python
sns.pairplot(data=tips,
             hue='day')
```

```python
daysplot =sns.relplot(data=tips,
                      x="total_bill",
                      y="tip",
                      hue="sex",
                      col="day",
                      kind="scatter",
                      col_wrap=2)
```

```python
import plotly.graph_objects as go

x1= np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1= np.array([110,180,240,99,220])
```

```python
graph=go.Figure()
graph.add_trace(go.Bar(x=x1,y=y1))
graph.update_layout(title="Pirate Scores",
                    xaxis_title ="Pirates",
                    yaxis_title="Score")
```

```python
graph=go.Figure()
graph.add_trace(go.Scatter(
    x=x1,
    y=y1,
    mode='markers',
    marker= dict(
        size=15,
        color='hotpink',
        opacity=1,
        line=dict(width=5,
        color='purple')
    )
))

graph.update_layout(title='Interactive Pirate Plot',
                    xaxis_title='Pirates',
                    yaxis_title='Scores',
                    width=500,
                    height=500)
```

```python

from wordcloud import WordCloud

df= pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
on_bad_lines ='skip')

df
```

```python
text = " ".join(each for each in df.quote)

wordcloud= WordCloud(background_color="white",
                     colormap='inferno').generate(text)
fig,ax=plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,
interpolation='bilinear')
ax.axis("off")
```
