1. Bar Chart: Number of notifications

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the data set 
data = pd.read_csv('/Users/72xs/visualization/02_activities/assignments/Tenant Notification for Rent Reduction 2023.csv')
print(data.columns)
data.columns =data.columns.str.strip()
print(data.head())
```

```python

#Filter only rows where notification were mailed 
data_cleaned = data[data["Tenant Notices Mailed (Y/N)"]== "Y"]

#Group data by "Political Ward" and count the number of notifications
ward_summary = data_cleaned.groupby('Political Ward').size().reset_index(name='Notification Count')
```

```python
#plot the bar chart 
plt.figure(figsize=(5, 3))
plt.bar(ward_summary['Political Ward'], ward_summary['Notification Count'], color='skyblue')
plt.title('Number of Tenant Notifications by Political Ward', fontsize=16)
plt.xlabel('Political Ward', fontsize=14)
plt.ylabel('Number of Notifications', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

2.Pie chart - proportion of notifications mailed vs. not mailed 

```python
# Count the values for mailed and not mailed
mailed_counts = data['Tenant Notices Mailed (Y/N)'].value_counts()

# Plot the pie chart
mailed_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
plt.title('Proportion of Notifications Mailed vs. Not Mailed', fontsize=16)
plt.ylabel('')  # Removes default y-axis label
plt.show()
```
